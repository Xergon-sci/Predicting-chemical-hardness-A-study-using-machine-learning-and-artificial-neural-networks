The plot below displays the spread of the data. For each value of hardness, the number of molecules are counted.

```
# Count the ammount that each hardness value exists.
hist_data_series = df['n'].value_counts()

# Proces the data to a dataframe
hist_data = hist_data_series.to_frame()
hist_data = hist_data.reset_index()
hist_data.columns = ['n','c']

# Sort the data on acsending hardness
hist_data = hist_data.sort_values(by=['n'])

# Plot the dat
plot.scatter(hist_data['n'], hist_data['c'])
plot.title('Spread of hardness')
plot.xlabel('Hardness')
plot.ylabel('Number of molecules')
plot.ylim(0,50)
plot.rcParams['figure.figsize'] = (14,6)
plot.show()
```

The plots below give the number of
